Inicialmente se realizará un reemplazo de los códigos que no corresponden a cada columna, posteriormente se cargará un dataset que contenga los códigos DIVIPOLA de departamentos y municipios para hacer una unión con el dataframe principal (df_casos_covid_colombia).

```python
# Unir dataframes

# Selecciona las primeras 1123 filas y todas las columnas del DataFrame df_codigos_DIVIPOLA
df_codigos_DIVIPOLA = df_codigos_DIVIPOLA.iloc[:1123,:]
# Convierte las columnas 'Código Departamento' y 'Código Municipio' a tipo entero de 64 bits
df_codigos_DIVIPOLA[['Código Departamento', 'Código Municipio']] = df_codigos_DIVIPOLA[['Código Departamento', 'Código Municipio']].astype('int64')


# Crea un nuevo DataFrame que contiene solo las columnas 'Código Departamento' y 'Nombre Departamento' del DataFrame original,
# eliminando las filas duplicadas para tener una lista única de códigos y nombres de departamentos.
df_codigos_DIVIPOLA_departamento = df_codigos_DIVIPOLA[['Código Departamento', 'Nombre Departamento']].drop_duplicates()
# Realiza una combinación (merge) del DataFrame df_casos_covid_colombia con df_codigos_DIVIPOLA_departamento.
# La combinación se realiza a la izquierda (left join), utilizando 'Código DIVIPOLA departamento' del primer DataFrame
# y 'Código Departamento' del segundo DataFrame como claves de unión.
df_casos_covid_colombia = pd.merge(df_casos_covid_colombia, df_codigos_DIVIPOLA_departamento, how='left', 
                                   left_on='Código DIVIPOLA departamento', right_on='Código Departamento')


# Mismo procedimiento para los códigos DIVIPOLA de los municipios
df_codigos_DIVIPOLA_municipio = df_codigos_DIVIPOLA[['Código Municipio', 'Nombre Municipio']].drop_duplicates()
df_casos_covid_colombia = pd.merge(df_casos_covid_colombia, df_codigos_DIVIPOLA_municipio, how='left', 
                                   left_on='Código DIVIPOLA municipio', right_on='Código Municipio')
```

### Corrección/Imputación de los valores nulos en las columnas `FECHA_DE_INICIO_DE_SINTOMAS` y `FECHA_DE_DIAGNOSTICO`

```python
# Cantidad de registros que tienen la misma fecha para ambas columnas
df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'][df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'] 
                                                       == df_casos_covid_colombia['FECHA_DE_DIAGNOSTICO']].value_counts().sum()
```

Un total de $274473$ casos tienen la misma fecha para ambas columnas, lo cual nos muestra que la **fecha de inicio de sintomas** y la **fecha diagnostico**, la cual es la fecha de confirmación por parte del laboratorio, coinciden en algunos casos.

```python
# Cantidad de valores nulos en 'FECHA_DE_INICIO_DE_SINTOMAS' cuando 'FECHA_DE_DIAGNOSTICO' no es nula
df_casos_covid_colombia[['FECHA_DE_INICIO_DE_SINTOMAS', 'FECHA_DE_DIAGNOSTICO']][df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'].isnull() &
                                                                                 df_casos_covid_colombia['FECHA_DE_DIAGNOSTICO'].notnull()]
```

Además, se observa que del total de valores nulos ($505384$) de la columna `FECHA_DE_INICIO_DE_SINTOMAS`, $505282$ registros presentan valores en la variable `FECHA_DE_DIAGNOSTICO`. Dado que esta última representa la fecha de confirmación por laboratorio, se asumirá que la prueba se realizó por sospecha el mismo día. Por lo tanto, se procederá a llenar los valores nulos de `FECHA_DE_INICIO_DE_SINTOMAS` con las fechas presentes en la variable `FECHA_DE_DIAGNOSTICO`.

```python
# Llenar los valores nulos de `FECHA_DE_INICIO_DE_SINTOMAS` con las fechas presentes en la variable `FECHA_DE_DIAGNOSTICO`
df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'] = df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'].fillna(df_casos_covid_colombia['FECHA_DE_DIAGNOSTICO'])
```

Para el resto de registros ($102$), que son nulos para ambas columnas, se toma la decisión de elimnarlas dado que representa apenas el $0.001\%$ de los datos.

```python
# Filtrar el DataFrame para encontrar las filas donde ambas columnas son nulas
eliminar_filas_na = df_casos_covid_colombia['FECHA_DE_INICIO_DE_SINTOMAS'].isnull() & df_casos_covid_colombia['FECHA_DE_DIAGNOSTICO'].isnull()

# Eliminar esas filas del DataFrame
df_casos_covid_colombia = df_casos_covid_colombia[~eliminar_filas_na]
```

Ahora verificaremos de nuevo los datos nulos del dataframe completo con los cambios realizados
